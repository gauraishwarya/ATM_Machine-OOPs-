# Creating Class

```python
class Atm:
    
    #creating Constructor
    def __init__(self): 
        self.pin = ""
        self.balance = 0
        self.menu() #calling Method "menu"

    #Defining Method "Menu"
    def menu(self):
        user_input = input(
                      """Hello,how would you like to proceed?
                        1. Enter 1 to create pin
                        2. Enter 2 to deposit
                        3. Enter 3 to withdraw
                        4. Enter 4 to check balance
                        5. Enter 5 to exit
                         Enter Option:- """
                        )
        if user_input == "1":
            self.create_pin()     #calling Method "create_pin"
        elif user_input == "2":
            self.deposit()        #calling Method "deposit"
        elif user_input == "3":
            self.withdraw()       #calling Method "withdraw"
        elif user_input == "4":
            self.check_balance()  #calling Method "check_balance"
        else:
            print("bye")
            
    
    #Defining Method "create_pin" 
    def create_pin(self):
        self.pin = input("Enter your pin:- ")
        print("Pin set successfully")
        self.main_menu()       #calling Method "main_menu"
    
    #Defining Method "deposit"
    def deposit(self):
        temp = input("Enter your pin:- ")
        if temp == self.pin:
            amount = int(input("Enter the amount to deposit:- "))
            self.balance = self.balance+amount
            print("Deposit Successful")
        else:
            print("Invalid pin")
        self.main_menu()       #calling Method "main_menu"
    
    #Defining Method "withdraw"
    def withdraw(self):
        temp = input("Enter your pin:- ")
        if temp == self.pin:
            amount = int(input("Enter the amount to withdraw:- "))
            if amount <= self.balance:
                self.balance = self.balance-amount
                print("Operation Successful")
            else:
                print("Insufficient Funds")
        else:
            print("Invalid pin")
        self.main_menu()      #calling Method "main_menu"

    #Defining Method "check_balance"
    def check_balance(self):
        temp = input("Enter your pin:- ")
        if temp == self.pin:
            print(self.balance)
        else:
            print("Invalid pin")
        self.main_menu()     #calling Method "main_menu"
    
    #Defining Method "main_menu"
    def main_menu(self):
        user_input = input(" Enter 0 to return to main menu else press any key to exit:- ")
        print()
        if user_input == "0":
            self.menu()     #calling Method "menu"
        else:
            print("Bye")
```
